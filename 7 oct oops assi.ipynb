```python
Q.1.............
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    # Method to deposit money into the account
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited {amount}. New balance: {self.balance}")
        else:
            print("Deposit amount must be greater than zero.")
    
    # Method to withdraw money from the account
    def withdraw(self, amount):
        if amount <= 0:
            print("Withdrawal amount must be greater than zero.")
        elif amount > self.balance:
            print("Insufficient funds for this withdrawal.")
        else:
            self.balance -= amount
            print(f"Withdrew {amount}. New balance: {self.balance}")
    
    # Method to display account details
    def display_account_info(self):
        print(f"Account Number: {self.account_number}")
        print(f"Account Holder: {self.account_holder_name}")
        print(f"Balance: {self.balance}")

# Example usage:
# Create a bank account instance
account = BankAccount(123456, "Pooja Rathore", 1000)

# Display initial account info
account.display_account_info()

# Deposit money
account.deposit(500)

# Withdraw money
account.withdraw(200)

# Withdraw more than the balance
account.withdraw(1500)

# Display updated account info
account.display_account_info()
```

```python
#Q.2...
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    # Method to calculate yearly bonus (assuming a 10% bonus of the salary)
    def calculate_bonus(self):
        bonus = self.salary * 0.10  # 10% bonus
        return bonus
    
    # Method to display employee details
    def display_employee_info(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Employee Name: {self.name}")
        print(f"Employee Salary: {self.salary}")
        print(f"Yearly Bonus: {self.calculate_bonus()}")

# Example usage:
# Create an employee instance
employee = Employee(101, "John Doe", 50000)

# Display employee information
employee.display_employee_info()
```

```python
# Q.4.........
class VehicleRental:
    def __init__(self):
        # List to store available vehicles and rented vehicles
        self.available_vehicles = ["Car", "SUV", "Truck", "Bike"]
        self.rented_vehicles = []

    # Method to rent a vehicle
    def rent_vehicle(self, vehicle):
        if vehicle in self.available_vehicles:
            self.available_vehicles.remove(vehicle)
            self.rented_vehicles.append(vehicle)
            print(f"Successfully rented the {vehicle}.")
        else:
            print(f"Sorry, {vehicle} is not available for rent.")
    
    # Method to return a vehicle
    def return_vehicle(self, vehicle):
        if vehicle in self.rented_vehicles:
            self.rented_vehicles.remove(vehicle)
            self.available_vehicles.append(vehicle)
            print(f"Thank you for returning the {vehicle}.")
        else:
            print(f"Error: {vehicle} was not rented.")
    
    # Method to display available vehicles
    def display_available_vehicles(self):
        if self.available_vehicles:
            print("Available vehicles for rent:")
            for vehicle in self.available_vehicles:
                print(f"- {vehicle}")
        else:
            print("No vehicles available for rent.")

# Example usage:
# Create a VehicleRental instance
rental_system = VehicleRental()

# Display available vehicles
rental_system.display_available_vehicles()

# Rent a vehicle
rental_system.rent_vehicle("Car")

# Display available vehicles after renting one
rental_system.display_available_vehicles()

# Return a vehicle
rental_system.return_vehicle("Car")

# Display available vehicles after returning one
rental_system.display_available_vehicles()

# Try to return a vehicle that wasn't rented
rental_system.return_vehicle("Truck")
```

```python
#Q.4............
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_borrowed = False

    # Method to display book details
    def display_book_info(self):
        status = "Borrowed" if self.is_borrowed else "Available"
        print(f"Title: {self.title}, Author: {self.author}, ISBN: {self.isbn}, Status: {status}")

class Library:
    def __init__(self):
        self.books = []

    # Method to add books to the library
    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' added to the library.")

    # Method to borrow a book
    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and not book.is_borrowed:
                book.is_borrowed = True
                print(f"You have successfully borrowed '{title}'.")
                return
        print(f"Sorry, the book '{title}' is either not available or already borrowed.")

    # Method to return a borrowed book
    def return_book(self, title):
        for book in self.books:
            if book.title == title and book.is_borrowed:
                book.is_borrowed = False
                print(f"Thank you for returning '{title}'.")
                return
        print(f"Error: You did not borrow the book '{title}'.")

    # Method to display available books
    def display_available_books(self):
        available_books = [book for book in self.books if not book.is_borrowed]
        if available_books:
            print("Available books in the library:")
            for book in available_books:
                book.display_book_info()
        else:
            print("No books are currently available for borrowing.")

# Example usage:

# Create a Library instance
library = Library()

# Create Book instances
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "9780743273565")
book2 = Book("1984", "George Orwell", "9780451524935")
book3 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084")

# Add books to the library
library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Display available books
library.display_available_books()

# Borrow a book
library.borrow_book("1984")

# Display available books after borrowing
library.display_available_books()

# Return a book
library.return_book("1984")

# Display available books after returning
library.display_available_books()

# Try to borrow a book that is already borrowed
library.borrow_book("1984")

# Try to return a book that was not borrowed
library.return_book("The Great Gatsby")
```

```python
#Q.5.
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    # Method to update product quantity
    def update_quantity(self, quantity):
        if quantity >= 0:
            self.quantity = quantity
            print(f"Updated quantity of '{self.name}' to {self.quantity}.")
        else:
            print("Quantity cannot be negative.")
    
    # Method to display product details
    def display_product_info(self):
        print(f"Product ID: {self.product_id}, Name: {self.name}, Price: {self.price}, Quantity: {self.quantity}")

class Inventory:
    def __init__(self):
        self.products = []

    # Method to add a product to the inventory
    def add_product(self, product):
        self.products.append(product)
        print(f"Product '{product.name}' added to inventory.")

    # Method to update the quantity of a product
    def update_product_quantity(self, product_id, quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.update_quantity(quantity)
                return
        print(f"Product with ID {product_id} not found in the inventory.")

    # Method to display available products
    def display_available_products(self):
        if self.products:
            print("Available products in the inventory:")
            for product in self.products:
                product.display_product_info()
        else:
            print("No products in the inventory.")

# Example usage:

# Create an Inventory instance
inventory = Inventory()

# Create Product instances
product1 = Product(101, "Laptop", 1000, 10)
product2 = Product(102, "Smartphone", 500, 20)
product3 = Product(103, "Headphones", 100, 50)

# Add products to the inventory
inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

# Display available products
inventory.display_available_products()
```

```python
#Q.6.............
class Shape:
    def __init__(self, length, width, height=0):
        self.length = length
        self.width = width
        self.height = height

    # Method to calculate the area of the shape
    def calculate_area(self):
        # If height is given, calculate the surface area (for a 3D shape like a cuboid)
        if self.height > 0:
            area = 2 * (self.length * self.width + self.length * self.height + self.width * self.height)
        else:
            # If height is not given, calculate area for a 2D rectangle
            area = self.length * self.width
        return area

    # Method to calculate the perimeter of the shape (only for 2D)
    def calculate_perimeter(self):
        if self.height == 0:
            # Perimeter of a rectangle (2D)
            perimeter = 2 * (self.length + self.width)
        else:
            # Perimeter of a cuboid (assuming we are interested in the perimeter of the base rectangle)
            perimeter = 4 * (self.length + self.width)
        return perimeter

    # Method to display shape details
    def display_shape_info(self):
        if self.height == 0:
            print(f"Shape: Rectangle, Length: {self.length}, Width: {self.width}")
            print(f"Area: {self.calculate_area()}")
            print(f"Perimeter: {self.calculate_perimeter()}")
        else:
            print(f"Shape: Cuboid, Length: {self.length}, Width: {self.width}, Height: {self.height}")
            print(f"Surface Area: {self.calculate_area()}")
            print(f"Perimeter of Base: {self.calculate_perimeter()}")

# Example usage:

# Create a Rectangle (2D)
rectangle = Shape(5, 3)

# Create a Cuboid (3D)
cuboid = Shape(5, 3, 4)

# Display details of the rectangle
print("Rectangle Details:")
rectangle.display_shape_info()

# Display details of the cuboid
print("\nCuboid Details:")
cuboid.display_shape_info()
```

```python
#Q.7.....
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades  # grades should be a list of numerical values

    # Method to calculate the average grade
    def calculate_average_grade(self):
        if self.grades:  # Check if the grades list is not empty
            average = sum(self.grades) / len(self.grades)
            return round(average, 2)  # Return the average rounded to 2 decimal places
        else:
            return 0.0  # If no grades are provided, return 0

    # Method to display student details
    def display_student_info(self):
        print(f"Student ID: {self.student_id}")
        print(f"Student Name: {self.name}")
        print(f"Grades: {self.grades}")
        print(f"Average Grade: {self.calculate_average_grade()}")

# Example usage:

# Create Student instances
student1 = Student(1, "John Doe", [85, 90, 78, 92])
student2 = Student(2, "Jane Smith", [88, 91, 85, 87])

# Display student information
print("Student 1 Details:")
student1.display_student_info()

print("\nStudent 2 Details:")
student2.display_student_info()
```

```python
#Q.8...............
class Email:
    def __init__(self, sender, recipient, subject, body):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = body
        self.sent = False  # To track whether the email has been sent

    # Method to simulate sending an email
    def send_email(self):
        if not self.sent:
            self.sent = True
            print(f"Email sent from {self.sender} to {self.recipient} with subject: {self.subject}")
        else:
            print("This email has already been sent.")

    # Method to display email details
    def display_email_details(self):
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print(f"Body: {self.body}")
        print(f"Email Sent: {'Yes' if self.sent else 'No'}")

# Example usage:

# Create an Email instance
email1 = Email("john.doe@example.com", "jane.smith@example.com", "Meeting Reminder", "Don't forget about the meeting at 3 PM tomorrow.")
email2 = Email("alice.jones@example.com", "bob.brown@example.com", "Project Update", "The project deadline is next week.")

# Display email details
print("Email 1 Details:")
email1.display_email_details()

# Send email 1
email1.send_email()

# Display email details again to show that it's sent
print("\nEmail 1 Details after sending:")
email1.display_email_details()

# Try to send the same email again
email1.send_email()

# Send email 2
email2.send_email()

# Display email 2 details
print("\nEmail 2 Details:")
email2.display_email_details()
```

```python
#Q.9.....................
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []  # A list to store posts

    # Method to add a post to the profile
    def add_post(self, content):
        self.posts.append(content)
        print(f"Post added: {content}")

    # Method to display all posts
    def display_posts(self):
        if self.posts:
            print(f"\n{self.username}'s Posts:")
            for index, post in enumerate(self.posts, start=1):
                print(f"{index}. {post}")
        else:
            print(f"{self.username} has no posts yet.")

    # Method to search for posts by keyword
    def search_posts(self, keyword):
        found_posts = [post for post in self.posts if keyword.lower() in post.lower()]
        if found_posts:
            print(f"\nPosts containing the keyword '{keyword}':")
            for post in found_posts:
                print(post)
        else:
            print(f"No posts found containing the keyword '{keyword}'.")

# Example usage:

# Create an instance of SocialMediaProfile
profile = SocialMediaProfile("john_doe")

# Add posts
profile.add_post("Had a great day at the park!")
profile.add_post("Learning Python is fun!")
profile.add_post("Looking forward to the weekend.")

# Display all posts
profile.display_posts()

# Search for posts containing a keyword
profile.search_posts("Python")
profile.search_posts("weekend")

# Try searching for a non-existing keyword
profile.search_posts("vacation")
```

```python
#Q.10..................
from datetime import datetime

class ToDoList:
    def __init__(self):
        self.tasks = []  # List to store tasks
        self.due_dates = []  # List to store due dates for tasks
        self.completed = []  # List to track task completion status

    # Method to add a task
    def add_task(self, task, due_date):
        self.tasks.append(task)
        self.due_dates.append(due_date)
        self.completed.append(False)  # Mark task as not completed
        print(f"Task added: {task} (Due: {due_date})")

    # Method to mark a task as completed
    def mark_task_completed(self, task_index):
        if 0 <= task_index < len(self.tasks):
            if not self.completed[task_index]:
                self.completed[task_index] = True
                print(f"Task '{self.tasks[task_index]}' marked as completed.")
            else:
                print(f"Task '{self.tasks[task_index]}' is already completed.")
        else:
            print("Invalid task index.")

    # Method to display pending tasks
    def display_pending_tasks(self):
        print("\nPending Tasks:")
        pending_found = False
        for index, (task, due_date, completed) in enumerate(zip(self.tasks, self.due_dates, self.completed)):
            if not completed:
                print(f"{index + 1}. {task} (Due: {due_date})")
                pending_found = True
        if not pending_found:
            print("All tasks are completed.")

# Example usage:

# Create an instance of ToDoList
todo_list = ToDoList()

# Add tasks
todo_list.add_task("Finish the report", "2025-01-15")
todo_list.add_task("Buy groceries", "2025-01-20")
todo_list.add_task("Call the plumber", "2025-01-18")

# Display pending tasks
todo_list.display_pending_tasks()

# Mark a task as completed
todo_list.mark_task_completed(1)

# Display pending tasks again
todo_list.display_pending_tasks()

# Try marking a task that is already completed
todo_list.mark_task_completed(1)

# Mark another task as completed
todo_list.mark_task_completed(0)

# Display pending tasks after completing some
todo_list.display_pending_tasks()
```
