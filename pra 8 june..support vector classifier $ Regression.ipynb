```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
y
```

```python
pd.DataFrame(x)[0]
pd.DataFrame(x)[1]
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.svm import SVC
```

```python
classifier=SVC(kernel='linear')
```

```python
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
classifier.coef_
```

```python
y_pred=classifier.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'C':[1,2,3,4,5,10,50,100],
    'gamma':[0.2,1.04,0.001,0.003],
    'kernel':['linear']
}
```

```python
params
```

```python
 grid=GridSearchCV(SVC(),param_grid=params,cv=5,verbose=3)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_regression
```

```python
x,y=make_regression(n_samples=1000,n_features=2,n_targets=1,noise=0.3)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.svm import SVR
svr=SVR(kernel='linear')
```

```python
svr.fit(x_train,y_train)
```

```python
y_pred=svr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    'C':[1,2,3,4,7,50,100],
    'gamma':[0.1,0.2,0.001,0.04],
    'kernel':['linear'],
    'epsilon':[0.1,0.001,0.2,0.3]
}
```

```python
grid=GridSearchCV(SVR(),param_grid=params,cv=5,verbose=3)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```
